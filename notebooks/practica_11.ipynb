## Operador IN

### Crea un subquery para generar una vista con:

**1 Los vendedores que no vendieron en 2018**

```sql
USE base;
SELECT * 
FROM vendedor
WHERE ID_Vendedor NOT IN(
    SELECT DISTINCT ID_Venta
    FROM ventas
    WHERE YEAR(STR_TO_DATE(`Fecha venta`, '%d-%b-%y')) = 2018);
```

**2 Productos que no se vendieron en marzo del 2017**

```sql
SELECT * 
FROM producto
WHERE ID_Producto NOT IN(
    SELECT DISTINCT ID_producto
    FROM ventas
    WHERE YEAR(STR_TO_DATE(`Fecha venta`, '%d-%b-%y')) = 2017
    AND MONTH(STR_TO_DATE(`Fecha venta`, '%d-%b-%y')) = 3);
```

**3. Clientes que compraron propiedas de remates Bancarios en el primer cuarto de 2016**

```sql

SELECT * 
FROM clientes
WHERE ID_Cliente IN(
    SELECT ID_producto
    FROM ventas
    WHERE ID_producto IN(
        SELECT ID_Producto 
        FROM producto
        WHERE Producto LIKE 'Remates'
            AND YEAR(STR_TO_DATE(`Fecha venta`, '%d-%b-%y')) = 2016
            AND MONTH(STR_TO_DATE(`Fecha venta`, '%d-%b-%y')) IN (1, 2, 3)

                        )
                    );
```

### <span style="font-size: 16.38px;">Uso de ALL &amp; ANY</span>

### <span style="font-size: 16.38px;">Genera una vista para entregar el informe que contenga:</span>

**4. Todas las ventas superiores a la venta más cara de una oficina**

\* **Todas las ventas superiores a la venta más cara de la Zona 1**

```sql
SELECT * 
FROM ventas 
WHERE Venta > ALL(
    SELECT Venta
    FROM ventas
    WHERE ID_Zona = 1);
```

**5. Cualquier venta mayor al de una oficina**

```sql
SELECT * 
FROM ventas 
WHERE Venta > ANY(
    SELECT Venta
    FROM ventas
    WHERE ID_Zona = 1
                );
```

## EXISTS

## Buscaremos a todos los vendedores que no vendieron usando EXISTS y el uso de un query correlacionado:

**6. Todos los vendedores que no han vendido nada**

```sql
SELECT * 
FROM vendedor vd 
WHERE NOT EXISTS(
    SELECT Venta 
    FROM ventas
    WHERE ID_Vendedor = vd.ID_Vendedor 
);
```

## Subqueries en SELECT

## Crea un subquery en SELECT para:

**7. Obtener el % que representa cada venta del total por cliente**

```sql
USE base;
SELECT 
    v.ID_Venta,
    v.ID_Cliente,
    v.Venta,
    (v.Venta / total_ventas.TotalCliente) * 100 AS Porcentaje
FROM 
    ventas v
JOIN 
    (SELECT 
         ID_Cliente, 
         SUM(Venta) AS TotalCliente 
     FROM 
         ventas 
     GROUP BY 
         ID_Cliente) total_ventas
ON 
    v.ID_Cliente = total_ventas.ID_Cliente;



```
